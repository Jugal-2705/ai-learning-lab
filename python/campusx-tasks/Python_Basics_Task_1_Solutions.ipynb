### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}C is equal to {fahrenheit}F")
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
coord1 = input("Enter x1 and y1 separated by a space: ").split()
x1 = float(coord1[0])
y1 = float(coord1[1])

coord2 = input("Enter x2 and y2 separated by a space: ").split()
x2 = float(coord2[0])
y2 = float(coord2[1])

distance = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
print("The Euclidean distance is:", distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
values = input("Enter Principal, Rate, and Time separated by spaces: ").split()
p = float(values[0])
r = float(values[1])
t = float(values[2])

simple_interest = (p * r * t) / 100
print("The Simple Interest is:", simple_interest)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
data = input("Enter total heads and legs separated by a space: ").split()
heads = int(data[0])
legs = int(data[1])

dogs = (legs - 2 * heads) / 2
chicken = heads - dogs

print("dogs ->", int(dogs))
print("chicken ->", int(chicken))
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
n = int(input("Enter the value of n: "))
sum_of_squares = (n * (n + 1) * (2 * n + 1)) / 6
print(f"The sum of squares of first {n} natural numbers is:", int(sum_of_squares))
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
data = input("Enter first term, common difference, and N separated by spaces: ").split()
a = float(data[0])
d = float(data[1])
n = int(data[2])
nth_term = a + (n - 1) * d
print(f"The {n}th term of the series is:", nth_term)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
frac1 = input("Enter first numerator and denominator separated by a space: ").split()
num1 = int(frac1[0])
den1 = int(frac1[1])

frac2 = input("Enter second numerator and denominator separated by a space: ").split()
num2 = int(frac2[0])
den2 = int(frac2[1])

numerator = (num1 * den2) + (num2 * den1)
denominator = den1 * den2

print(f"The sum of the fractions is {numerator}/{denominator}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
import math
tank = input("Enter tank Height, Length, and Breadth separated by spaces: ").split()
H = float(tank[0])
L = float(tank[1])
B = float(tank[2])

glass = input("Enter glass Height and Radius separated by a space: ").split()
h = float(glass[0])
r = float(glass[1])

tank_volume = H * L * B
glass_volume = math.pi * (r**2) * h

glasses = tank_volume / glass_volume
print(f"Total glasses of milk that can be obtained: {int(glasses)}")
````
